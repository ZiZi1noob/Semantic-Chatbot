```python
def Text_clean(text):
    text=re.sub(r' \n|\n', ' ', text)
    text = re.sub(r'http\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'@\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'#', '', text, flags=re.MULTILINE)
    text = re.sub(r'，', '', text, flags=re.MULTILINE)
    text = re.sub(r'！', '', text, flags=re.MULTILINE)
    text = re.sub(r'？', '', text, flags=re.MULTILINE)
    text = re.sub(r'。', '', text, flags=re.MULTILINE)
    text = re.sub(r'“', '', text, flags=re.MULTILINE)
    text = re.sub(r'”', '', text, flags=re.MULTILINE)
    text = re.sub(r'、', '', text, flags=re.MULTILINE)
    text=re.sub('[%s]' % re.escape(string.punctuation), ' ',text)
    text = re.sub('\w*\d\w*', '', text)
    return text

round=lambda x:Text_clean(x)
```

```python
tw['text']=pd.DataFrame(tw['text'].apply(round))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier
from sklearn.model_selection import train_test_split
```

```python
tw_copy=tw.copy()
```

```python
train,test = train_test_split(tw_copy,test_size=0.2,random_state=42)
```

```python
tw_train=[]
for i in train['text']:
    tw_train.append(i)
tw_test=[]
for i in test['text']:
    tw_test.append(i)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv_class = CountVectorizer()
train_features= cv_class.fit_transform(tw_train)
test_features=cv_class.transform(tw_test)
```

```python
Classifiers = [
    LogisticRegression(),
    GaussianNB(),
    DecisionTreeClassifier(max_depth=1),
    RandomForestClassifier(n_estimators=200)]
```

```python
Classifiers = [
    LogisticRegression(),
    GaussianNB(),
    DecisionTreeClassifier(max_depth=1),
    RandomForestClassifier(n_estimators=200)]
```

```python
dense_features=train_features.toarray()
dense_test= test_features.toarray()
Accuracy=[]
Model=[]
for classifier in Classifiers:
    try:
        fit = classifier.fit(train_features,train['Sentiment'])
        pred = fit.predict(test_features)
    except Exception:
        fit = classifier.fit(dense_features,train['Sentiment'])
        pred = fit.predict(dense_test)
    accuracy = accuracy_score(pred,test['Sentiment'])
    Accuracy.append(accuracy)
    Model.append(classifier.__class__.__name__)

for i in range(len(Accuracy)):
    print (Model[i],": ",Accuracy[i])
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
logreg = LogisticRegression(solver='liblinear')
logreg.fit(train_features, train['Sentiment'])
pred = logreg.predict(test_features)
print('The Micro Precision score is',precision_score(test['Sentiment'], pred, average='micro'),'\n')
#print(confusion_matrix(test['Sentiment'],pred))
print(classification_report(test['Sentiment'],pred))
```
